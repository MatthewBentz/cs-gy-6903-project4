```python
# 1. Why hash collisions are inevitable
# A hash function takes any input and gives a fixed-size output. 
# Since there are more possible inputs than outputs, two different inputs will eventually produce the same hash. 
# This is called a collision, and it is unavoidable.
```

```python
# 2. How Merkle trees work
# A Merkle tree is a structure that uses hashes to organize data.
# Each file (or data block) is hashed → these are the leaf nodes
# Pairs of hashes are combined and hashed again
# This continues until one final hash is created, called the root hash
# If any data changes, the root hash also changes, which helps detect tampering.
```

```python
# 3. How proof-of-work functions in blockchain systems
# Proof-of-work is a way to secure a blockchain by requiring computers to solve hard puzzles.
# Miners try to find a number (nonce) that produces a valid hash
# This takes a lot of computation
# Once found, others can easily verify it
# This process helps keep the network secure and prevents cheating.
```
