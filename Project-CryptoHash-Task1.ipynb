```python
import hashlib
import os
```

```python
def sha256(data):
    return hashlib.sha256(data.encode('utf-8')).hexdigest()
```

```python
def read_file(file_path):
    with open(file_path, 'r') as f:
        return f.read()
```

```python
def build_merkle_tree(file_paths):
    # Step 1: Compute leaf hashes
    leaves = [sha256(read_file(f)) for f in file_paths]
    
    tree = [leaves]  # store all levels
    
    # Step 2: Build tree upward
    current_level = leaves
    
    while len(current_level) > 1:
        next_level = []
        
        for i in range(0, len(current_level), 2):
            left = current_level[i]
            
            # handle odd number of nodes
            if i + 1 < len(current_level):
                right = current_level[i + 1]
            else:
                right = left
            
            parent = sha256(left + right)
            next_level.append(parent)
        
        tree.append(next_level)
        current_level = next_level
    
    return tree
```

```python
def print_tree(tree):
    for level_num, level in enumerate(tree):
        print(f"Level {level_num}:")
        for node in level:
            print(node)
        print()
```

```python
files_4 = ["file1.txt", "file2.txt", "file3.txt", "file4.txt"]

contents = ["Hello", "Merkle", "Tree", "Example"]

for f, c in zip(files_4, contents):
    with open(f, 'w') as file:
        file.write(c)
```

```python
tree_4 = build_merkle_tree(files_4)
print("Merkle Tree with 4 leaves:\n")
print_tree(tree_4)

root_4 = tree_4[-1][0]
print("Root Hash:", root_4)
```

```python
files_6 = ["f1.txt", "f2.txt", "f3.txt", "f4.txt", "f5.txt", "f6.txt"]

contents_6 = ["A", "B", "C", "D", "E", "F"]

for f, c in zip(files_6, contents_6):
    with open(f, 'w') as file:
        file.write(c)
```

```python
tree_6 = build_merkle_tree(files_6)
print("Merkle Tree with 6 leaves:\n")
print_tree(tree_6)

root_6 = tree_6[-1][0]
print("Root Hash:", root_6)
```

```python
# 1. When one leaf changes, which nodes change?
# When a leaf changes:
# Its hash changes
# All parent nodes up to the root change
# Only the path from that leaf to the root is affected.
```

```python
# 2. Why does a small change produce a completely different root hash?
# Because of the avalanche effect:
# Even a tiny input change produces a drastically different hash output
```

```python
# 3. What properties make Merkle trees useful?
# Deterministic: same input → same hash
# Collision-resistant: hard to find two inputs with same hash
# Avalanche effect: small changes → big differences
# Efficient verification: only part of the tree needed to verify data
```

```python
# 4. Real-world systems using Merkle trees
# Blockchain (e.g., Bitcoin, Ethereum)
# Used to verify transactions efficiently
# Git
# Uses hash trees to track file changes and ensure integrity
```
