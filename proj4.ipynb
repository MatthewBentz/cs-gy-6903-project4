## Part I. Merkle Trees and Data Integrity

### Task

Implement a binary Merkle tree using a cryptographic hash function such as SHA-256. Each leaf node of the tree should correspond to a plaintext file. The hash of each internal node is computed as the hash of the concatenation of its two children. 

### Experiments

1. Construct a Merkle tree using four leaf files and display the resulting tree structure and hashes. 

```python
import os
import hashlib

# Helper functions

def merkle_hash(left, right):
    return hashlib.sha256((left + right).encode()).hexdigest()

def generate_files(path, amount):
    os.makedirs(path, exist_ok=True)
    for i in range(amount):
        with open(path + f"/file{i}.txt", "w") as f:
            f.write(f"This is file {i} at {path}.\n\nHere is a random string: {os.urandom(16).hex()}\n")

def get_file_hashes(path, amount):
    hashes = {}
    for i in range(amount):
        with open(path + f"/file{i}.txt", "rb") as f:
            data = f.read()
            hash_value = hashlib.sha256(data).hexdigest()
            hashes[i] = hash_value
    return hashes
```

```python
path = "part1/four_leaf"
amount = 4
hashes = {}

# Generate 4 files for the Merkle tree

generate_files(path, amount)

# Hash each file

hashes = get_file_hashes(path, amount)

# Generate/display the Merkle tree

# Level 1 (leaves)
hash0 = hashes[0]
hash1 = hashes[1]
hash2 = hashes[2]
hash3 = hashes[3]
# Level 2
hash0_1 = merkle_hash(hash0, hash1)
hash2_3 = merkle_hash(hash2, hash3)
# Level 3 (root)
root_hash = merkle_hash(hash0_1, hash2_3)
print("Merkle Tree:")
print(f"\tRoot Hash: {root_hash}")
print(f"\t\tHash0_1: {hash0_1}")
print(f"\t\t\tHash0: {hash0}")
print(f"\t\t\tHash1: {hash1}")
print(f"\t\tHash2_3: {hash2_3}")
print(f"\t\t\tHash2: {hash2}")
print(f"\t\t\tHash3: {hash3}")
```

```python
path = "part1/six_leaf"
amount = 6
hashes = {}

# Generate 6 files for the Merkle tree

generate_files(path, amount)

# Hash each file

hashes = get_file_hashes(path, amount)

# Generate the Merkle tree

# Level 1 (leaves)
hash0 = hashes[0]
hash1 = hashes[1]
hash2 = hashes[2]
hash3 = hashes[3]
hash4 = hashes[4]
hash5 = hashes[5]
# Level 2
hash0_1 = merkle_hash(hash0, hash1)
hash2_3 = merkle_hash(hash2, hash3)
hash4_5 = merkle_hash(hash4, hash5)
# Level 3
hash0_1_2_3 = merkle_hash(hash0_1, hash2_3)
# Level 4 (root)
root_hash = merkle_hash(hash0_1_2_3, hash4_5)
print("Merkle Tree:")
print(f"\tRoot Hash: {root_hash}")
print(f"\t\tHash0_1_2_3: {hash0_1_2_3}")
print(f"\t\t\tHash0_1: {hash0_1}")
print(f"\t\t\t\tHash0: {hash0}")
print(f"\t\t\t\tHash1: {hash1}")
print(f"\t\t\tHash2_3: {hash2_3}")
print(f"\t\t\t\tHash2: {hash2}")
print(f"\t\t\t\tHash3: {hash3}")
print(f"\t\tHash4_5: {hash4_5}")
print(f"\t\t\tHash4: {hash4}")
print(f"\t\t\tHash5: {hash5}")
```

3. Modify the content of one of the files used in the four-leaf example. Recompute the tree and compare the new root hash with the original one.

### Questions

1. When the content of one leaf file is modified, which nodes in the tree change?

In this example, 3 nodes in the tree change. The hash of the original file that was modified, the has of file0/file1, and the root hash. In the above output, you can compare the Root Hash, Hash0_1, and Hash0.

2. Why does a small change in the file lead to a completely different root hash?

The small change in a file leads to a completely different root hash because of the avalanche effect properties in hashing functions, such as SHA-256 as used in the experiment. The drastic change in a leaf node will cascade recursively updwards, causing the root hash and any of it's derived children to the leaf to be significantly changed.

3. What properties of cryptographic hash functions make Merkle trees useful?

As described, the avalanche effect makes Merkle trees useful due to the drastic change in output. When a small change is reflected throughout the tree, you are able to traverse from the root to see exactly which file changed, but also have high confidence in the validity of the tree since the effect compounds. Additionally, collision resistance helps in this confidence, since its infeasible that a change will result in the same hash (and therefore, the same root).

4. Identify at least two real-world systems that make use of Merkle trees and briefly describe their purpose.

Two real-world systems that use Merkle trees would be blockchains (such as crypto-currencies, i.e. Bitcoin) and Git. In the Bitcoin example, each block is represented by a Merkle tree, and its purpose is to provide a lightweight and verifiable representation of the entire transaction list in the block. Additionally, the Merkle tree structure serves to allow $O(log_n)$ searches, rather than $O(n)$. For the Git example, Merkle trees allow commits to store the entire state of the repository by a hash (the trees root hash) at the specific point in time. Because we can detect at which level in the tree the commits changed, we can also efficiently identify the files that have changed in the repo between commits.

## Part III. Hash Puzzles and Proof-of-Work

### Task

Construct a simple hash puzzle of the following form: compute hash(message || nonce) and search for a nonce such that the resulting hash beings with a specified number of zero bits.

### Experiments

Solve puzzles requiring:

1. one leading zero bit

2. two leading zero bits

3. three leading zero bits

Record the number of attempts required in each case.

```python
import hashlib

message = "hello"
nonce = 0
count = 0  #this counter tracks number of attempts

while True:
    text = message + str(nonce)

    hash_result = hashlib.sha256(text.encode()).hexdigest()

    count += 1  #increment counter

    # Leading Zero Bits- Adjust leading zeros as needed
    if hash_result.startswith("00"):
        print("Nonce found:", nonce)
        print("Hash:", hash_result)
        print("Attempts:", count)
        break

    nonce += 1
```
